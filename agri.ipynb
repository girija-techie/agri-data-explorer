```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Read csv file
df = pd.read_csv('./../Agri_Data.csv')
df
```

```python
duplicates = df[df.duplicated()]
print("total duplicates rows:",len(duplicates))
```

```python
# Data cleaning

df = df.drop(["Dist Code","State Code","BARLEY AREA (1000 ha)","BARLEY PRODUCTION (1000 tons)","BARLEY YIELD (Kg per ha)","CHICKPEA AREA (1000 ha)","CHICKPEA PRODUCTION (1000 tons)","CHICKPEA YIELD (Kg per ha)","PIGEONPEA AREA (1000 ha)","PIGEONPEA PRODUCTION (1000 tons)","PIGEONPEA YIELD (Kg per ha)","MINOR PULSES AREA (1000 ha)","MINOR PULSES PRODUCTION (1000 tons)","MINOR PULSES YIELD (Kg per ha)","FRUITS AREA (1000 ha)","VEGETABLES AREA (1000 ha)","FRUITS AND VEGETABLES AREA (1000 ha)","POTATOES AREA (1000 ha)","ONION AREA (1000 ha)","FODDER AREA (1000 ha)","SESAMUM AREA (1000 ha)","SESAMUM PRODUCTION (1000 tons)","SESAMUM YIELD (Kg per ha)","RAPESEED AND MUSTARD AREA (1000 ha)","RAPESEED AND MUSTARD PRODUCTION (1000 tons)","RAPESEED AND MUSTARD YIELD (Kg per ha)","SAFFLOWER AREA (1000 ha)","SAFFLOWER PRODUCTION (1000 tons)","SAFFLOWER YIELD (Kg per ha)","CASTOR AREA (1000 ha)","CASTOR PRODUCTION (1000 tons)","CASTOR YIELD (Kg per ha)","LINSEED AREA (1000 ha)","LINSEED PRODUCTION (1000 tons)","LINSEED YIELD (Kg per ha)"], axis=1)
```

```python
df
```

```python
# Exploratory Data Analysis (EDA) Questions and Answers

# 1.Top 7 RICE PRODUCTION State Data(Bar_plot)

df_rice = df[["State Name","Year", "RICE PRODUCTION (1000 tons)"]]
```

```python
state_rice = df_rice.groupby("State Name")["RICE PRODUCTION (1000 tons)"].sum().reset_index()
```

```python
top7_rice_states = state_rice.sort_values(by="RICE PRODUCTION (1000 tons)", ascending=False).head(7)
```

```python
plt.Figure(figsize=(12,6))
sns.barplot(data=top7_rice_states, x="RICE PRODUCTION (1000 tons)", y ="State Name" , palette= "plasma")
plt.title("Top 7 Rice Production States in India", fontsize=14)
plt.xlabel("Total Rice Production(1000 tons)")
plt.ylabel("States")
plt.show()
```

```python
# 2. Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)

df_wheat = df[["State Name", "Year", "WHEAT PRODUCTION (1000 tons)"]]
```

```python
state_wheat = df_wheat.groupby("State Name")["WHEAT PRODUCTION (1000 tons)"].sum().reset_index()
```

```python
top5_wheat_states = state_wheat.sort_values(by="WHEAT PRODUCTION (1000 tons)", ascending=False).head(5)
```

```python
top5_wheat_states["Percentage"] = (top5_wheat_states["WHEAT PRODUCTION (1000 tons)"]/top5_wheat_states["WHEAT PRODUCTION (1000 tons)"].sum())*100
```

```python
# Top 5 Wheat Producing States Data(Bar_chart)

plt.Figure(figsize=(12, 6))
sns.barplot(data=top5_wheat_states, x="WHEAT PRODUCTION (1000 tons)", y="State Name", palette="inferno")
plt.title("Top 5 Wheat Production States in india", fontsize=14)
plt.xlabel("Total Wheat Production (1000 tons)")
plt.ylabel("State")
plt.show()
```

```python
# Top 5 Wheat Producing States Data(Pie_chart)

plt.figure(figsize=(5,5))
plt.pie(top5_wheat_states["Percentage"],labels = top5_wheat_states["State Name"],autopct="%.1f%%",startangle = 140)
plt.title("Percentage of Top 5 Wheat Production States in India",fontsize=12)
plt.show()
```

```python
# 3. Oil seed production by top 5 states

df_oil = df[["State Name", "Year", "OILSEEDS PRODUCTION (1000 tons)"]]
```

```python
state_oil = df_oil.groupby("State Name")["OILSEEDS PRODUCTION (1000 tons)"].sum().reset_index()
```

```python
top5_oil_states = state_oil.sort_values(by="OILSEEDS PRODUCTION (1000 tons)", ascending=False).head(5)
```

```python
# Line Chart

plt.figure(figsize=(10, 6))
sns.lineplot(data=top5_oil_states, x="OILSEEDS PRODUCTION (1000 tons)", y ="State Name" , color= "Green", marker = "o")
plt.title("Top 5 Oilseeds Producing States in India",fontsize=14)
plt.xlabel("Total Oilseeds Production(1000 tons)")
plt.ylabel("State")
plt.show()
```

```python
# 4. Top 7 SUNFLOWER PRODUCTION State

df_sunflower = df[["State Name", "Year", "SUNFLOWER PRODUCTION (1000 tons)"]]
```

```python
state_sunflower = df_sunflower.groupby("State Name")["SUNFLOWER PRODUCTION (1000 tons)"].sum().reset_index()
```

```python
top7_sunflower_states = state_sunflower.sort_values(by="SUNFLOWER PRODUCTION (1000 tons)", ascending=False).head(7)
```

```python
# 5. India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)

print("Earliest Year:", df["Year"].min())
print("Latest Year:", df["Year"].max())
```

```python
df_sugarcane = df[["State Name", "Year", "SUGARCANE PRODUCTION (1000 tons)"]]
india_sugarecane = df_sugarcane.groupby("Year")["SUGARCANE PRODUCTION (1000 tons)"].sum().reset_index()
india_sugarecane_last50year = india_sugarecane.tail(50)
```

```python
# Line Plot

plt.figure(figsize=(12, 6))
plt.plot(india_sugarecane_last50year["Year"],india_sugarecane_last50year["SUGARCANE PRODUCTION (1000 tons)"], marker="o", color="purple", linewidth=2)
plt.title("India's Sugarcane Production (Last 50 Years)",fontsize=14)
plt.xlabel("Year")
plt.ylabel("Sugarcane Production (1000 tons)")
plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```

```python
# 6. Rice Production Vs Wheat Production (Last 50y)

df_rice_vs_wheat = df[["Year", "RICE PRODUCTION (1000 tons)", "WHEAT PRODUCTION (1000 tons)"]]
rice_vs_wheat = df.groupby("Year")[["RICE PRODUCTION (1000 tons)", "WHEAT PRODUCTION (1000 tons)"]].sum().reset_index()
rice_vs_wheat_last50year = rice_vs_wheat.sort_values("Year").tail(50)
```

```python
# Stacked Area Chart

plt.figure(figsize=(12, 6))
plt.plot(rice_vs_wheat_last50year["Year"],rice_vs_wheat_last50year["RICE PRODUCTION (1000 tons)"],marker="o",color="green",linewidth=2,label="Rice")
plt.plot(rice_vs_wheat_last50year["Year"],rice_vs_wheat_last50year["WHEAT PRODUCTION (1000 tons)"],marker="o",color="gold",linewidth=2,label="Wheat")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.title("Rice vs Wheat Production in India (Last 50 Years)")
plt.legend()
plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```

```python
# 7. Rice Production By West Bengal Districts

df_rice_westbengal = df[(df["State Name"]=="West Bengal")][["Dist Name", "Year", "RICE PRODUCTION (1000 tons)"]]
rice_westbengal = df_rice_westbengal.groupby("Dist Name")["RICE PRODUCTION (1000 tons)"].sum().reset_index()
rice_westbengal_dist = rice_westbengal.sort_values("RICE PRODUCTION (1000 tons)", ascending=False)
```

```python
# Donut Chart
plt.figure(figsize=(7,7))
plt.pie(rice_westbengal_dist["RICE PRODUCTION (1000 tons)"], labels=rice_westbengal_dist["Dist Name"], autopct="%1.1f%%", startangle=210, pctdistance=0.85, colors=plt.cm.Paired.colors)

# Draw circle in middle (to make donut)
centre_circle = plt.Circle((0,0), 0.70, fc="white")
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Rice Production in West Bengal Districts")
plt.tight_layout()
plt.show()
```

```python
# 8. Top 10 Wheat Production Years From UP

df_wheat_up = df[df["State Name"] == "Uttar Pradesh"][["Year", "WHEAT PRODUCTION (1000 tons)"]]
wheat_up = df_wheat_up.groupby("Year")["WHEAT PRODUCTION (1000 tons)"].sum().reset_index()
top10_wheat_up = wheat_up.sort_values("WHEAT PRODUCTION (1000 tons)", ascending=False).head(10)
```

```python
# Bar chart

colors = ["red", "gold", "purple", "green", "lightblue", "violet", "blue", "grey", "orange", "pink"]
plt.figure(figsize=(10,6))
bars = plt.bar(top10_wheat_up["Year"].astype(str), top10_wheat_up["WHEAT PRODUCTION (1000 tons)"], color=colors[:len(top10_wheat_up)])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, 
             f"{int(height):,}",  # formatted with commas
             ha="center", va="bottom", fontsize=9, fontweight="bold")

plt.xlabel("Year")
plt.ylabel("Wheat Production (1000 tons)")
plt.title("Year wise Top 10 Wheat Production in Uttar Pradesh")
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Millet Production (Last 50y)

df_millet = df.groupby("Year")[["SORGHUM PRODUCTION (1000 tons)", "RABI SORGHUM PRODUCTION (1000 tons)", "KHARIF SORGHUM PRODUCTION (1000 tons)", "PEARL MILLET PRODUCTION (1000 tons)", "FINGER MILLET PRODUCTION (1000 tons)"]].sum().reset_index()
df_millet_last50yr = df_millet.sort_values("Year").tail(50)
```

```python
#  Stacked Bar Chart
df_millet_last50yr.set_index("Year")[["SORGHUM PRODUCTION (1000 tons)", "RABI SORGHUM PRODUCTION (1000 tons)", "KHARIF SORGHUM PRODUCTION (1000 tons)", "PEARL MILLET PRODUCTION (1000 tons)", "FINGER MILLET PRODUCTION (1000 tons)"]].plot(kind="bar", stacked=True, figsize=(14, 6), colormap="tab10")

plt.title("Last 50 Years Millet Production in India")
plt.ylabel("Millet Production (1000 tons)")
plt.xlabel("Year")
plt.legend(title="Millet Type")
plt.show()
```

```python
# 10. Sorghum Production (Kharif and Rabi) by Region

df_sorghum_region = df.groupby("State Name")[["KHARIF SORGHUM PRODUCTION (1000 tons)", "RABI SORGHUM PRODUCTION (1000 tons)"]].sum().reset_index()
```

```python
# 11. Top 7 States for Groundnut Production

df_groundnut = df.groupby("State Name")[["GROUNDNUT PRODUCTION (1000 tons)"]].sum().reset_index()
top7_states_groundnut = df_groundnut.sort_values("GROUNDNUT PRODUCTION (1000 tons)", ascending=False).head(7)
```

```python
# 12. Soybean Production by Top 5 States and Yield Efficiency

df_soyabean = df.groupby("State Name")[["SOYABEAN PRODUCTION (1000 tons)", "SOYABEAN AREA (1000 ha)"]].sum().reset_index()

# Soyabean Yeild Efficiency

df_soyabean["Yeild Efficiency (tons/ha)"] = df_soyabean["SOYABEAN PRODUCTION (1000 tons)"]/df_soyabean["SOYABEAN AREA (1000 ha)"]

top5_soyabean_states = df_soyabean.sort_values("SOYABEAN PRODUCTION (1000 tons)", ascending=False).head(5)
```

```python
# Soyabean production Line Chart 

plt.figure(figsize=(10, 6))
sns.lineplot(data=top5_soyabean_states, x="SOYABEAN PRODUCTION (1000 tons)", y ="State Name" , color= "purple", marker = "o")
plt.title("Top 5 Soyabean Production States in India",fontsize=14, color="darkgreen")
plt.xlabel("Total Soyabean Production(1000 tons)")
plt.ylabel("State Name")
plt.show()
```

```python
# Top 5 states Soyabean Yeild Efficiency 

x = np.arange(len(top5_soyabean_states["State Name"]))
width = 0.35  

plt.figure(figsize=(10,6))

plt.bar(x + width/2, top5_soyabean_states["Yeild Efficiency (tons/ha)"], width, label="Yield Efficiency", color="lightgreen")
plt.xticks(x, top5_soyabean_states["State Name"], rotation=30, ha="right")
plt.ylabel("Yeild Value(tons/ha)")
plt.xlabel("States")
plt.title("Top 5 States Soyabean Yield Efficiency")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# # 14. Impact of Area Cultivated on Production (Rice, Wheat, Maize)

df_area_production = df.groupby("Year")[[ "RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)","WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)","MAIZE PRODUCTION (1000 tons)", "MAIZE AREA (1000 ha)"]].sum().reset_index()
```

```python
# Regression Line Plot
plt.figure(figsize=(12,6))
sns.regplot(x="RICE AREA (1000 ha)", y="RICE PRODUCTION (1000 tons)", data=df_area_production, scatter_kws={'s':30}, label="Rice", color="#37f056")
sns.regplot(x="WHEAT AREA (1000 ha)", y="WHEAT PRODUCTION (1000 tons)", data=df_area_production, scatter_kws={'s':30}, label="Wheat", color="#f0a626")
sns.regplot(x="MAIZE AREA (1000 ha)", y="MAIZE PRODUCTION (1000 tons)", data=df_area_production, scatter_kws={'s':30}, label="Maize", color="#f026c8")

plt.xlabel("Area Cultivated (1000 ha)", fontsize=12)
plt.ylabel("Production (1000 tons)", fontsize=12)
plt.title("Correlation between Area and Production", fontsize=15 )
plt.legend()
plt.show()
```

```python
# 15. Rice vs. Wheat Yield Across States

df_yield = df.groupby("State Name")[["RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)", "WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)"]].sum().reset_index()

df_yield["Rice Yield"] = df_yield["RICE PRODUCTION (1000 tons)"] / df_yield["RICE AREA (1000 ha)"]

df_yield["Wheat Yield"] = df_yield["WHEAT PRODUCTION (1000 tons)"] / df_yield["WHEAT AREA (1000 ha)"]

df_yield_rice_wheat = df_yield[["State Name", "Rice Yield", "Wheat Yield"]]
```

```python
# Heatmap Plot

data_heat = df_yield_rice_wheat.set_index("State Name")[["Rice Yield","Wheat Yield"]]

plt.figure(figsize=(10,10))
sns.heatmap(data_heat, annot=True, cmap="inferno", fmt=".3f")
plt.title("Rice vs Wheat Yield Across States in India", fontsize= 14)
plt.xlabel("Yield Production of Rice/Wheat", fontsize= 12)
plt.show()
```
